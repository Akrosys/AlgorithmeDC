```
a = 2
b = 10
c = 25

delta = b*b - 4*a*c

if delta > 0:
  x1 = (-b - sqrt(delta))/(2*a)
  x2 = (-b + sqrt(delta))/(2*a)
  print("Les racines de l'équation sont : x1 = ",x1, " et x2 = ", x2 )
elif delta == 0:
  x1 = -b/(2*a)
  print("La racine de l'équation est : x = ", x1)
else:
  print("Les racines de l'équation sont : x1 = (b - i*sqrt(-delta))/2*a et x2 = (b + i*sqrt(-delta))/2 * a")
```

```
fac = int(input("Entrez un nombre :"))
res = 1
i = 1
while i <= fac:
  res *= i
  i += 1
print("La factorielle de ", fac, " est : ", res)
```

```
n = int(input("Entrez la valeur de n :"))
k = int(input("Entrez la valeur de k :"))

def fact(n):
  x = 1
  for i in range(2, n+1):
    x *= i
  return x

CoeffBinomial = fact(n) / (fact(k) * fact(n - k))
print(CoeffBinomial)
```

```
import numpy as np

def creationMatrice():
  mat = []
  lignes = int(input("Lignes : "))
  colonnes = int(input("Colonnes : "))
  mat = np.zeros((lignes,colonnes))

  for j in range(0, colonnes):
    for i in range(0,lignes):
      mat[i,j] = input("Valeur : ")
  return mat
print("Entrez la matrice A :")
A = creationMatrice()
print("Entrez la matrice B :")
B = creationMatrice()

print(np.dot(A,B))
```
